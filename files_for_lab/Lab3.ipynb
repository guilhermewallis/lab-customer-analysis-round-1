```python
import pandas as pd
```

1. Read the three files into python as dataframes

```python
data1 = pd.read_csv("csv_files/file1.csv")
data2 = pd.read_csv("csv_files/file2.csv")
data3 = pd.read_csv("csv_files/file3.csv")
```

2. Show the DataFrame's shape.

```python
print('data1', data1.shape)
print('data2', data2.shape)
print('data3', data3.shape)
```

3. Standardize header names

```python
data1.columns = [col.lower().replace(' ', '_') for col in data1.columns]
data2.columns = [col.lower().replace(' ', '_') for col in data2.columns]
data3.columns = [col.lower().replace(' ', '_') for col in data3.columns]
data3
```

4. Rearrange the columns in the dataframe as needed

```python
print(data1.columns)
print(data2.columns)
print(data3.columns)

data2 = data2[list(data1.columns.values)]
data3.rename(columns = {'state':'st'}, inplace = True)
data3 = data3[list(data1.columns.values)]
```

5. Concatenate the three dataframes

```python
data = pd.concat([data1, data2, data3])
```

6. Which columns are numerical?
7. Which columns are categorical?

```python
display(data.select_dtypes('number'))
#income 	monthly_premium_auto 	total_claim_amount
display(data.select_dtypes('object'))
#customer 	st 	gender 	education 	customer_lifetime_value 	number_of_open_complaints 	policy_type 	vehicle_class
```

8. Understand the meaning of all columns

```python
print(data.columns)
```

Columns:
customer - customer ID
st - US state of residence
gender - customer's gender
education - level of education of customer
customer_lifetime_value - total worth of customer 
income - customer's yearly income
monthly_premium_auto - monthly insurance cost for customer

policy_type - type of policy (personal/corporate/special)
vehicle_class - type of vehicle customer owns
total_claim_amount - total amount paid to customer in claims so far

9. Perform the data cleaning operations mentioned so far in class

```python
#Delete the column education and the number of open complaints from the dataframe.

data = data.drop(['education', 'number_of_open_complaints'], axis=1)
data
```

```python
#Correct the values in the column customer lifetime value.

data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace(r'[^0-9]+', '')
data['customer_lifetime_value'] = data['customer_lifetime_value'].astype(float)
data['customer_lifetime_value'] = data['customer_lifetime_value'].fillna(0)
data['customer_lifetime_value'] = data['customer_lifetime_value'] * 100
data['customer_lifetime_value']
```

```python
#Check for duplicate rows in the data and remove if any.

print(data.shape) #(12074, 9)
dups = data[data.duplicated()]
print(dups.shape) #(2943, 9)
data = data.drop_duplicates()
print(data.shape) #(9131, 9)
```

```python
#Filter out the data for customers who have an income of 0 or less.

print(data.shape) #(9131, 9)
data = data[data['income'] > 0]
print(data.shape) #(6837, 9)
```
